```python
import os
from shutil import copyfile
data_path = 'server/larger_data'
```

```python
os.makedirs('bism_data', exist_ok = True)
for file in os.listdir(data_path):
    if 'بسم' in file:
        copyfile(f'{data_path}/{file}', f'bism_data/{file}')
```
